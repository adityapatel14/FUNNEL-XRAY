```python
stages = ["Login", "View", "Click", "Add to Cart", "Purchase"]
users = [800, 601, 421, 253, 139]
```

```python
funnel_df = pd.DataFrame({"Stage": stages, "Users": users})
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=funnel_df, x="Users", y="Stage", palette="Blues_d")
plt.title("SaaS User Funnel - Drop off Analysis")
plt.xlabel("Number of Users")
plt.ylabel("Funnel Stage")
plt.show()
```
